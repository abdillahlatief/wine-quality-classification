# **4. Pembuatan Model**

4.1 Memisah prediktor dan target

```python
# Memisahkan variabel prediktor (X) dan target (y)
X = train.drop(columns=['Id', 'quality'])
y = train['quality']

# Menyiapkan data testing untuk prediksi akhir
X_test = test.drop(columns=['Id'])

# Melihat dimensi data
print("X shape      :", X.shape)
print("y shape      :", y.shape)
print("X_test shape :", X_test.shape)
```

4.2 Strategi Evaluasi Model

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

# membuat skema cross validation
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

print(cv)
```

4.3 Pelatihan dan Evaluasi Model Random Forest

```python
rf = RandomForestClassifier(
    n_estimators=300,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf_scores = cross_val_score(
    rf,
    X,
    y,
    cv=cv,
    scoring='accuracy',
    n_jobs=-1
)

print("Random Forest CV Scores :", rf_scores)
print("Mean Accuracy           :", rf_scores.mean())
print("Std Dev                 :", rf_scores.std())
```

4.4 Pelatihan dan Evaluasi Model ExtraTrees

```python
et = ExtraTreesClassifier(
    n_estimators=300,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

et_scores = cross_val_score(
    et,
    X,
    y,
    cv=cv,
    scoring='accuracy',
    n_jobs=-1
)

print("ExtraTrees CV Scores :", et_scores)
print("Mean Accuracy        :", et_scores.mean())
print("Std Dev              :", et_scores.std())
```

4.5 Pelatihan dan Evaluasi Model XGBoost

```python
from sklearn.preprocessing import LabelEncoder

# encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Class asli   :", sorted(y.unique()))
print("Class encode :", sorted(np.unique(y_encoded)))
```

```python
xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    objective='multi:softmax',
    num_class=6,
    random_state=42,
    n_jobs=-1
)

xgb_scores = cross_val_score(
    xgb,
    X,
    y_encoded,
    cv=cv,
    scoring='accuracy',
    n_jobs=-1
)

print("XGBoost CV Scores :", xgb_scores)
print("Mean Accuracy     :", xgb_scores.mean())
print("Std Dev           :", xgb_scores.std())
```

4.6 Perbandingan Performa Model

```python
comparison = pd.DataFrame({
    'Model': ['Random Forest', 'ExtraTrees', 'XGBoost'],
    'Mean Accuracy': [
        rf_scores.mean(),
        et_scores.mean(),
        xgb_scores.mean()
    ],
    'Std Dev': [
        rf_scores.std(),
        et_scores.std(),
        xgb_scores.std()
    ]
})

comparison = comparison.sort_values(
    by='Mean Accuracy',
    ascending=False
)

comparison
```

```python
print("RF  :", rf_scores)
print("ET  :", et_scores)
print("XGB :", xgb_scores)

print("\nMean RF  :", rf_scores.mean())
print("Mean ET  :", et_scores.mean())
print("Mean XGB :", xgb_scores.mean())
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x='Model',
    y='Mean Accuracy'
)

plt.ylim(
    comparison['Mean Accuracy'].min() - 0.01,
    comparison['Mean Accuracy'].max() + 0.01
)

plt.title('Perbandingan Akurasi Model')
plt.show()
```

4.7 Pemilihan Model Terbaik

Jika dilihat dari visualisasi sebelumnya, maka model terbaiknya adalah Random Forest dengan mean accuracy-nya adalah: **0.64649122**


4.8 Pelatihan Model Final

```python
best_model = RandomForestClassifier(
    n_estimators=300,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

best_model.fit(X, y)
```

4.9 Prediksi Data Testing

```python
pred = best_model.predict(X_test)

print(pred[:20])
print("Jumlah prediksi:", len(pred))
```

4.10 Export ke CSV

```python
submission = pd.DataFrame({
    'Id': test['Id'],
    'quality': pred
})

submission.to_csv(
    'hasilprediksi.csv',
    index=False
)

submission.head()
```

4.11 Optimasi Hyperparameter Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV
param_dist = {
    'n_estimators': [200, 300, 500],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}
rf_tuned = RandomForestClassifier(
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

random_search = RandomizedSearchCV(
    estimator=rf_tuned,
    param_distributions=param_dist,
    n_iter=20,
    cv=cv,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1,
    verbose=1
)

random_search.fit(X, y)
```

```python
print("Best Params:")
print(random_search.best_params_)

print("\nBest Score:")
print(random_search.best_score_)
```
